```python
import pandas as pd

# Cargar únicamente la columna 'Review_scores_accuracy' del archivo CSV
file_path = "../0_Dataset/Listings.csv"
data = pd.read_csv(file_path, usecols=['property_type','price'], delimiter=',', encoding='latin-1')

# Verificar si la columna 'Review_scores_accuracy' existe en el DataFrame
if 'property_type' in data.columns:
    # Tipo de dato
    tipo_dato = data['property_type'].dtype
    
    # Cantidad de nulos
    cantidad_nulos = data['property_type'].isna().sum()
    
    # Cantidad de vacíos (considerando cadenas vacías)
    cantidad_vacios = (data['property_type'] == '').sum()
    
    # Cantidad de datos únicos
    cantidad_unicos = data['property_type'].nunique()
    
    # Cantidad de datos totales
    cantidad_totales = data['property_type'].shape[0]
    
    # Imprimir resultados
    print(f"Columna 'property_type':")
    print(f" - Tipo de dato: {tipo_dato}")
    print(f" - Cantidad de nulos: {cantidad_nulos}")
    print(f" - Cantidad de vacíos: {cantidad_vacios}")
    print(f" - Cantidad de datos únicos: {cantidad_unicos}")
    print(f" - Cantidad de datos totales: {cantidad_totales}")
else:
    print("La columna 'property_typee' no existe en el DataFrame.")

print(data['property_type'].value_counts())
print(data['property_type'].unique())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar únicamente la columna 'property_type' del archivo CSV
file_path = "../0_Dataset/Listings.csv"
data = pd.read_csv(file_path, usecols=['property_type'], delimiter=',', encoding='latin-1')

# Verificar si la columna 'property_type' existe en el DataFrame
if 'property_type' in data.columns:
    # Contar la frecuencia de cada valor
    value_counts = data['property_type'].value_counts()
    porcentajes = (value_counts / value_counts.sum()) * 100
    
    # Identificar categorías que representan menos del 1%
    categorias_poco_frecuentes = porcentajes[porcentajes < 1].index
    
    # Crear una copia del DataFrame original
    data_modificado = data.copy()
    
    # Reemplazar las categorías poco frecuentes por 'Otros'
    data_modificado.loc[data_modificado['property_type'].isin(categorias_poco_frecuentes), 'property_type'] = 'Otros'
    
    # Calcular las nuevas frecuencias y porcentajes exactos
    new_value_counts = data_modificado['property_type'].value_counts()
    new_porcentajes = (new_value_counts / new_value_counts.sum()) * 100
    
    # Ordenar de mayor a menor
    new_value_counts = new_value_counts.sort_values(ascending=False)
    new_porcentajes = new_porcentajes[new_value_counts.index]
    
    # Tu paleta de colores extendida
    base_colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A', '#FF689D']
    extended_colors = base_colors + [
        '#FFD166', '#06D6A0', '#118AB2', '#EF476F', '#073B4C',
        '#7209B7', '#F72585', '#3A86FF', '#8338EC', '#FF006E'
    ]
    
    # Crear el gráfico de tarta con porcentajes exactos
    fig, ax = plt.subplots(figsize=(12, 8))
    
    # Dibujar el pie con tus colores personalizados
    wedges, texts, autotexts = ax.pie(
        new_value_counts,
        startangle=30,
        colors=extended_colors[:len(new_value_counts)],
        wedgeprops={'linewidth': 1.5, 'edgecolor': 'white'},
        autopct=lambda p: f'{p:.2f}%' if p >= 1 else '',  # Mostrar 2 decimales para porcentajes exactos
        pctdistance=0.85,
        textprops={'fontsize': 10, 'color': 'white', 'weight': 'bold'}
    )
    
    # Añadir leyenda detallada a la derecha con porcentajes exactos
    legend_labels = [f"{nombre} ({new_porcentajes.loc[nombre]:.2f}%)" for nombre in new_value_counts.index]
    
    ax.legend(
        wedges,
        legend_labels,
        title="Tipos de propiedad",
        loc="center left",
        bbox_to_anchor=(1, 0, 0.5, 1),
        fontsize=10
    )
    
    ax.set_title('Distribución de tipo de propiedades', fontsize=16, pad=20)
    
    plt.tight_layout()
    plt.show()
    
    # Mostrar categorías agrupadas en 'Otros' con porcentajes exactos
    if len(categorias_poco_frecuentes) > 0:
        print(f"\nSe agruparon {len(categorias_poco_frecuentes)} categorías en 'Otros' (cada una <1%):")
        for cat, pct in porcentajes[categorias_poco_frecuentes].items():
            print(f"   {cat}: {pct:.2f}%")
else:
    print("La columna 'property_type' no existe en el DataFrame.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import numpy as np

# Cargar únicamente la columna 'property_type' del archivo CSV


# Verificar si la columna 'property_type' existe en el DataFrame
if 'property_type' in data.columns:
    # Contar la frecuencia de cada valor
    value_counts = data['property_type'].value_counts()
    
    # Crear un diccionario para el WordCloud con todas las categorías
    freq_dict = value_counts.to_dict()
    
    # Generar el WordCloud
    wordcloud = WordCloud(
        width=800, 
        height=400, 
        background_color='white',
        colormap='viridis',
        max_words=200,  # Aumentado para mostrar más categorías
        min_font_size=8,  # Reducido para permitir más palabras
        max_font_size=150,
        random_state=42
    ).generate_from_frequencies(freq_dict)
    
    # Visualizar el WordCloud
    plt.figure(figsize=(12, 8))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('WordCloud de Todos los Tipos de Propiedades', fontsize=16)
    plt.tight_layout()
    plt.show()
    
    # Mostrar información adicional
    print(f"WordCloud generado con {len(value_counts)} tipos de propiedades.")
    print(f"Categoría más frecuente: {value_counts.index[0]} ({value_counts.iloc[0]} ocurrencias)")
    print(f"Categoría menos frecuente: {value_counts.index[-1]} ({value_counts.iloc[-1]} ocurrencias)")
else:
    print("La columna 'property_type' no existe en el DataFrame.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Filtrar datos válidos (sin nulos y precios positivos)
data_valida = data[['property_type', 'price']].dropna()
data_valida = data_valida[data_valida['price'] > 0]

# 2. Obtener los 10 tipos de propiedad más frecuentes
top_tipos = data_valida['property_type'].value_counts().head(10).index

# 3. Filtrar solo esos tipos
data_filtrada = data_valida[data_valida['property_type'].isin(top_tipos)]

# 4. Calcular el precio promedio por tipo de propiedad
precio_promedio = data_filtrada.groupby('property_type')['price'].mean().sort_values(ascending=False)

# 5. Traducir nombres de tipos de propiedad
traduccion_tipos = {
   'Entire house': 'Casa completa',
    'Entire serviced apartment': 'Departamento con servicios',
    'Room in boutique hotel': 'Habitación en hotel boutique',
    'Condominium': 'Condominio',
    'Private room in bed and breakfast': 'Hab. privada en B&B',
    'Private room in condominium': 'Hab. privada en condominio',
    'Entire loft': 'Loft completo',
    'Private room in house': 'Hab. privada en casa',
    'Entire apartment': 'Departamento',
    'Private room in apartment': 'Hab. privada en dpto.',
    'Entire condominium': 'Condominio',
}
precio_promedio.index = [traduccion_tipos.get(tipo, tipo) for tipo in precio_promedio.index]

# 6. Graficar
colores = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FECA57',
           '#F67280', '#6C5B7B', '#C06C84', '#FFA07A', '#66CC99']

plt.figure(figsize=(12, 7))
ax = precio_promedio.plot(kind='bar', color=colores)

plt.title("Precio promedio por tipo de propiedad", fontsize=16, pad=20)
plt.ylabel("Precio promedio (USD)", fontsize=12)
plt.xlabel("Tipo de propiedad", fontsize=12)
plt.xticks(rotation=45, ha='right')

# Grilla y limpieza de bordes
ax.yaxis.grid(True, linestyle='--', alpha=0.5)
for lado in ['top', 'right']:
    ax.spines[lado].set_visible(False)

plt.tight_layout()
plt.show()
```
