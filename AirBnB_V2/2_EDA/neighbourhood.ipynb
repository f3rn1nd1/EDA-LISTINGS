```python
import pandas as pd

# Cargar únicamente la columna 'Review_scores_accuracy' del archivo CSV
file_path = "../0_Dataset/Listings.csv"
data = pd.read_csv(file_path, usecols=['neighbourhood','city'], delimiter=',', encoding='latin-1')

# Verificar si la columna 'Review_scores_accuracy' existe en el DataFrame
if 'neighbourhood' in data.columns:
    # Tipo de dato
    tipo_dato = data['neighbourhood'].dtype
    
    # Cantidad de nulos
    cantidad_nulos = data['neighbourhood'].isna().sum()
    
    # Cantidad de vacíos (considerando cadenas vacías)
    cantidad_vacios = (data['neighbourhood'] == '').sum()
    
    # Cantidad de datos únicos
    cantidad_unicos = data['neighbourhood'].nunique()
    
    # Cantidad de datos totales
    cantidad_totales = data['neighbourhood'].shape[0]
    
    # Imprimir resultados
    print(f"Columna 'neighbourhood':")
    print(f" - Tipo de dato: {tipo_dato}")
    print(f" - Cantidad de nulos: {cantidad_nulos}")
    print(f" - Cantidad de vacíos: {cantidad_vacios}")
    print(f" - Cantidad de datos únicos: {cantidad_unicos}")
    print(f" - Cantidad de datos totales: {cantidad_totales}")
else:
    print("La columna 'neighbourhoode' no existe en el DataFrame.")

print(data['neighbourhood'].value_counts())
print(data['neighbourhood'].unique())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Verificar si la columna 'neighbourhood' existe en el DataFrame
if 'neighbourhood' in data.columns:
    # Contar la frecuencia de cada valor
    value_counts = data['neighbourhood'].value_counts()
    porcentajes = (value_counts / value_counts.sum()) * 100
    
    # Identificar categorías que representan menos del 1.5%
    categorias_poco_frecuentes = porcentajes[porcentajes < 1.5].index
    
    # Crear una copia del DataFrame original
    data_modificado = data.copy()
    
    # Reemplazar las categorías poco frecuentes por 'Otros'
    data_modificado.loc[data_modificado['neighbourhood'].isin(categorias_poco_frecuentes), 'neighbourhood'] = 'Otros'
    
    # Calcular las nuevas frecuencias y porcentajes
    new_value_counts = data_modificado['neighbourhood'].value_counts()
    new_porcentajes = (new_value_counts / new_value_counts.sum()) * 100
    
    # Crear etiquetas con nombres y porcentajes
    etiquetas = [f"{categoria} ({porcentaje:.2f}%)" for categoria, porcentaje in zip(new_value_counts.index, new_porcentajes)]
    
    # Crear el gráfico de tarta
    fig, ax = plt.subplots(figsize=(10, 8))
    ax.pie(new_value_counts, labels=etiquetas, startangle=30)
    ax.set_title('Distribución de Vecindarios', fontsize=16)
    
    plt.tight_layout()
    plt.show()
    
    # Mostrar cuántas y cuáles categorías se agruparon en 'Otros'
    if len(categorias_poco_frecuentes) > 0:
        print(f"\nSe agruparon {len(categorias_poco_frecuentes)} categorías en 'Otros' (cada una <1.5%):")
        for cat, pct in porcentajes[categorias_poco_frecuentes].items():
            print(f"   {cat}: {pct:.2f}%")
else:
    print("La columna 'neighbourhood' no existe en el DataFrame.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import numpy as np

# Cargar únicamente la columna 'neighbourhood' del archivo CSV
file_path = "../0_Dataset/Listings.csv"
data = pd.read_csv(file_path, usecols=['neighbourhood'], delimiter=',', encoding='latin-1')

# Verificar si la columna 'neighbourhood' existe en el DataFrame
if 'neighbourhood' in data.columns:
    # Contar la frecuencia de cada valor
    value_counts = data['neighbourhood'].value_counts()
    
    # Crear un diccionario para el WordCloud con todas las categorías
    freq_dict = value_counts.to_dict()
    
    # Generar el WordCloud
    wordcloud = WordCloud(
        width=800, 
        height=400, 
        background_color='white',
        colormap='viridis',
        max_words=200,  # Aumentado para mostrar más categorías
        min_font_size=8,  # Reducido para permitir más palabras
        max_font_size=150,
        random_state=42
    ).generate_from_frequencies(freq_dict)
    
    # Visualizar el WordCloud
    plt.figure(figsize=(12, 8))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('WordCloud de Todos los Tipos de Propiedades', fontsize=16)
    plt.tight_layout()
    plt.show()
    
    # Mostrar información adicional
    print(f"WordCloud generado con {len(value_counts)} tipos de propiedades.")
    print(f"Categoría más frecuente: {value_counts.index[0]} ({value_counts.iloc[0]} ocurrencias)")
    print(f"Categoría menos frecuente: {value_counts.index[-1]} ({value_counts.iloc[-1]} ocurrencias)")
else:
    print("La columna 'neighbourhood' no existe en el DataFrame.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# OPCIÓN 1: Heatmap de concentración por ciudad
def heatmap_barrios_ciudades():
    # Top 10 ciudades y sus top 8 barrios
    top_cities = data['city'].value_counts().head(10).index
    
    # Crear matriz para el heatmap
    matriz_data = []
    city_labels = []
    
    for city in top_cities:
        top_barrios = data[data['city'] == city]['neighbourhood'].value_counts().head(8)
        matriz_data.append(top_barrios.values)
        city_labels.append(city)
    
    # Convertir a DataFrame
    max_len = max(len(row) for row in matriz_data)
    matriz_normalizada = []
    for row in matriz_data:
        normalized_row = list(row) + [0] * (max_len - len(row))
        matriz_normalizada.append(normalized_row)
    
    plt.figure(figsize=(16, 10))
    
    # Crear heatmap
    sns.heatmap(matriz_normalizada, 
                annot=True, 
                fmt='g',
                cmap='YlOrRd',
                xticklabels=[f'Barrio {i+1}' for i in range(max_len)],
                yticklabels=city_labels,
                cbar_kws={'label': 'Cantidad de Alojamientos'})
    
    plt.title('Concentración de Alojamientos por Vecindarios\nTop 10 Ciudades - Ordenados por Popularidad', 
              fontsize=16, pad=20, fontweight='bold')
    plt.xlabel('Ranking de Barrios (1 = Más Popular)', fontsize=12)
    plt.ylabel('Ciudades', fontsize=12)
    plt.tight_layout()
    plt.show()

# EJECUTAR TODAS LAS OPCIONES
print("=== OPCIÓN 1: Heatmap de Concentración ===")
heatmap_barrios_ciudades()
```

```python
import pandas as pd

def tabla_barrios_top_80(data):
    # Obtener las 80 ciudades con más alojamientos
    top_cities = data['city'].value_counts().head(80).index

    tabla_auxiliar = []

    for city in top_cities:
        top_barrios = data[data['city'] == city]['neighbourhood'].value_counts().head(8)
        for i, (barrio, cantidad) in enumerate(top_barrios.items(), start=1):
            tabla_auxiliar.append({
                'Ciudad': city,
                'Ranking': f'Barrio {i}',
                'Nombre del Barrio': barrio,
                'Cantidad de Alojamientos': cantidad
            })

    # Crear DataFrame auxiliar
    df_aux = pd.DataFrame(tabla_auxiliar)

    # Mostrar la tabla completa (sin truncar)
    pd.set_option('display.max_rows', None)
    print("\n=== Tabla Auxiliar: Top 8 Barrios por Ciudad (80 ciudades) ===")
    print(df_aux.to_string(index=False))

    return df_aux

# Ejecutar
tabla_auxiliar = tabla_barrios_top_80(data)
```
